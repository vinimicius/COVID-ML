```python
# 3. Importação da função mestra
from data_utils import run_pipeline

# 4. Execução do Teste Completo
try:
    print("🧪 Iniciando Teste de Integração...")
    
    # Rodamos com export=True para testar a nova função export_data
    df = run_pipeline(data_dir='data', export=True)
    
    print("\n--- Verificações de Sucesso ---")
    
    # Verificação 1: O objeto DataFrame está correto?
    if df is not None and not df.empty:
        print(f"✅ DataFrame em memória: {df.shape[0]:,} linhas.")
    
    # Verificação 2: O arquivo foi realmente salvo no disco?
    ready_path = 'data/data_ready.csv'
    if os.path.exists(ready_path):
        size_mb = os.path.getsize(ready_path) / (1024 * 1024)
        print(f"✅ Arquivo exportado encontrado: {ready_path} ({size_mb:.2f} MB)")
    
    # Verificação 3: Tipagem (O ponto crítico para ML)
    if df.dtypes.map(lambda x: x == 'int64' or x == 'int32').all():
        print("✅ Tipagem: Todas as colunas são numéricas (inteiros).")

except Exception as e:
    print(f"❌ O teste falhou miseravelmente: {e}")
```
